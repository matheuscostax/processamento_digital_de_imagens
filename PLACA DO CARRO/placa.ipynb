```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_path = '../PLACA DO CARRO/car_plate_1.png'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap='gray')
plt.title("Imagem em escala de cinza")
plt.axis('off')
plt.show()
```

```python
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.figure()
plt.title("Histograma da Imagem")
plt.xlabel("Intensidade de Pixels")
plt.ylabel("Nmero de Pixels")
plt.plot(hist)
plt.xlim([0, 256])
plt.show()
```

```python
threshold_value = 117
_, thresh_image = cv2.threshold(image, threshold_value, 255, cv2.THRESH_BINARY)
```

```python
plt.figure()
plt.imshow(thresh_image, cmap='gray')
plt.title(f'Imagem com Threshold = {threshold_value}')
plt.axis('off')
plt.show()
```

```python
kernel = np.ones((3, 3), np.uint8)
eroded_image = cv2.erode(thresh_image, kernel, iterations=2)
```

```python
plt.figure()
plt.imshow(eroded_image, cmap='gray')
plt.title('Imagem aps Eroso')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(image, cmap='gray')
plt.title("Imagem em escala de cinza")
plt.axis('off')

# Segundo subplot
plt.subplot(1, 3, 2)
plt.imshow(thresh_image, cmap='gray')
plt.title(f'Imagem com Threshold = {threshold_value}')
plt.axis('off')

# Terceiro subplot
plt.subplot(1, 3, 3)
plt.imshow(eroded_image, cmap='gray')
plt.title('Imagem aps Eroso')
plt.axis('off')

plt.show()
```
