FILTRO MORFOLOGICO

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
imagem = cv2.imread('../images/lena.png', cv2.IMREAD_GRAYSCALE)
_, imagem_binaria = cv2.threshold(imagem, 127, 255, cv2.THRESH_BINARY)
kernel = np.ones((3, 3), np.uint8)
imagem_erosao = cv2.erode(imagem_binaria, kernel, iterations=1)
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title("Imagem Original")
plt.imshow(imagem, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Imagem após Erosão")
plt.imshow(imagem_erosao, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
