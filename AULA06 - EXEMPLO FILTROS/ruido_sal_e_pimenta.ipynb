RUIDO SAL E PIMENTA

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def add_salt_pepper_noise(image, prob):
    noisy_image = image.copy()

    salt_mask = np.random.random(image.shape[:2]) < prob/2
    pepper_mask = np.random.random(image.shape[:2]) < prob/2

    noisy_image[salt_mask] = 255
    noisy_image[pepper_mask] = 0
    
    return noisy_image
```

```python
original_image = cv2.imread('../images/lena.png')
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
```

```python
noyse_image = add_salt_pepper_noise(original_image, 0.05)
plt.imshow(noyse_image)
plt.show()
```

```python
def remove_salt_peper_noise(image, kernel_size=3):
    if len(image.shape) ==3:
        result = np.zeros_like(image)
        for i in range(3):
            result[:,:,i] = cv2.medianBlur(image[:, :, i], kernel_size)
        else:
            result = cv2.medianBlur(image, kernel_size)

        return result
```

```python
denoised_image = remove_salt_peper_noise(noyse_image)
plt.imshow(denoised_image)
plt.show()
```

```python
plt.figure(figsize = (15, 5))

plt.subplot (1, 3, 1)
plt.imshow(original_image)
plt.title('Imagem Original')
plt.axis('off')

plt.subplot (1, 3, 2)
plt.imshow(noyse_image)
plt.title('Sal e Pimenta')
plt.axis('off')

plt.subplot (1, 3, 3)
plt.imshow(denoised_image)
plt.title('Imagem Filtrada')
plt.axis('off')
```
