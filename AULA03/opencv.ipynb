```python
import cv2    #opencv
import matplotlib.pyplot as plt   #matplotlib
import numpy as np      #numpy
```

```python
image_bgr = cv2.imread('../images/exemplo.png')
```

```python
def show_image(imagem):
    try:
        plt.imshow(image_bgr)
        plt.axis('off')
        plt.show()
    except TypeError:
        print("A imagem não é valida")
```

```python
show_image(image_bgr)
```

```python
#Tentativa de imagem rgb
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_RGB2BGR)
plt.imshow(image_rgb)
```

```python
image_gray = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
```

```python
red, green, blue = cv2.split (image_rgb)

blank = np.zeros(image_rgb.shape[:2], dtype='uint8')

canal_vermelho = cv2.merge([red, blank, blank])
canal_verde = cv2.merge([blank, green, blank])
canal_azul = cv2.merge([blank, blank, blue])
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 4, 1)
plt.imshow(canal_vermelho)
plt.title('Red')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(canal_verde)
plt.title('Green')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(canal_azul)
plt.title('Blue')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(image_rgb)
plt.title('Rgb')
plt.axis('off')



```

```python
inverted_image = 1 - (image_rgb / 255.0)
inverted_image = (inverted_image * 255).astype(np.uint8)
inverted_image
```
